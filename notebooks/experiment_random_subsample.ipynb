# Load the data

```python
from tabpfn_project.paths import EXPERIMENTS_DATA_DIR, RESULTS_DIR
from tabpfn_project.helper.utils import load_pickle, fetch_save_dict
from tabpfn_project.helper.utils import plot_instance_level_bar_chart
from tabpfn_project.paths import RESULTS_DIR

tabpfn_results_dir = RESULTS_DIR / "TabPFN"
distnet_results_dir = RESULTS_DIR / "DistNet"
rf_results_dir = RESULTS_DIR / "RandomForest"
lognormal_results_dir = RESULTS_DIR / "LogNormal"
gp_results_dir = RESULTS_DIR / "GaussianProcess"

tabpfn_log_results = load_pickle(tabpfn_results_dir / "tabpfn_results_exp_random_subsample_log.pkl")
tabpfn_max_results = load_pickle(tabpfn_results_dir / "tabpfn_results_exp_random_subsample_max.pkl")
tabpfn_log_noDUPS_results = load_pickle(tabpfn_results_dir / "tabpfn_results_exp_random_subsample_log_noDUPS.pkl")
tabpfn_naive_results = load_pickle(tabpfn_results_dir / "tabpfn_results_exp_random_subsample_naive.pkl")
tabpfn_oracle_results = load_pickle(tabpfn_results_dir / "tabpfn_oracle.pkl")

distnet_log_results = load_pickle(distnet_results_dir / "distnet_results_exp_random_subsample_max_log_target_scale_log.pkl")
distnet_max_results = load_pickle(distnet_results_dir / "distnet_results_exp_random_subsample_max_log_target_scale_max.pkl")

random_forest_results = load_pickle(rf_results_dir / "random_forest_results_exp_random_subsample_log.pkl")
random_forest_results_old = load_pickle(rf_results_dir / "random_forest_results_exp_random_subsample_log_old.pkl")
random_forest_results_tuned = load_pickle(rf_results_dir / "random_forest_results_exp_random_subsample_log_tuned.pkl")

lognormal_results = load_pickle(lognormal_results_dir / "lognormal_results_exp_random_subsample_naive.pkl")
lognormal_oracle_results = load_pickle(lognormal_results_dir / "lognormal_oracle.pkl")

gp_results = load_pickle(gp_results_dir / "gp_results_exp_random_subsample_log.pkl")
```
