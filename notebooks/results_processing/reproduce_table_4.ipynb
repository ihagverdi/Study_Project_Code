```python
import pickle
import pandas as pd
from pathlib import Path
import numpy as np

# Define the path to the data folder
data_folder = Path(r'c:\Users\ihagv\Desktop\masters_project\project_code\cluster\table_4_data\metadata')

# Dictionary to store results organized by scenario
scenario_results = {}

# Process all pickle files
for pkl_file in data_folder.glob('*.pkl'):
    try:
        # Load the pickle file
        with open(pkl_file, 'rb') as f:
            data = pickle.load(f)
        
        # Extract scenario and NLLH value
        scenario = data['config']['scenario']
        nllh = data['metrics']['nllh']
        
        # Store the result
        if scenario not in scenario_results:
            scenario_results[scenario] = []
        scenario_results[scenario].append(nllh)
        
    except Exception as e:
        print(f"Error processing {pkl_file.name}: {e}")

# Calculate mean NLLH for each scenario
mean_nllh = {}
for scenario, nllh_values in scenario_results.items():
    mean_nllh[scenario] = np.mean(nllh_values)
    print(f"{scenario}: {len(nllh_values)} folds found")

# Create a DataFrame
df = pd.DataFrame.from_dict(mean_nllh, orient='index', columns=['Mean NLLH'])
df.index.name = 'Scenario'
df = df.sort_index()

# Display the table
print("\n" + "="*50)
print("Table 4: Mean NLLH across 10 folds")
print("="*50)
print(df)

# Optional: Save to CSV
```

```python
df.to_csv('table_4_results.csv')
```
