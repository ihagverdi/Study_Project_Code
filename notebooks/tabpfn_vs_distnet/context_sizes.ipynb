```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import defaultdict

# ── 1. Load all .pkl files ────────────────────────────────────────────────────

PKL_DIR = r"C:\Users\ihagv\Desktop\Study Project UFR\Study_Project_Code\results\results_data\tabpfn_context\metadata"

records = []
for fname in os.listdir(PKL_DIR):
    if not fname.endswith(".pkl"):
        continue
    fpath = os.path.join(PKL_DIR, fname)
    with open(fpath, "rb") as f:
        d = pickle.load(f)
    records.append({
        "model_name":   d["model_name"],
        "scenario":     d["scenario"],
        "fold":         d["fold"],
        "context_size": d["context_size"],
        "nllh":         d["result_metrics"]["nllh"],
    })

df = pd.DataFrame(records)
print(f"Loaded {len(df)} records")
print("Scenarios :", sorted(df["scenario"].unique()))
print("Models    :", sorted(df["model_name"].unique()))
print("Context sizes:", sorted(df["context_size"].unique()))

# ── 2. Aggregation ────────────────────────────────────────────────────────────
fold_means = (
    df.groupby(["model_name", "scenario", "fold", "context_size"])["nllh"]
    .mean()
    .reset_index()
    .rename(columns={"nllh": "nllh_fold_mean"})
)

agg = (
    fold_means.groupby(["model_name", "scenario", "context_size"])["nllh_fold_mean"]
    .agg(mean="mean", std="std")
    .reset_index()
)

# ── 3. Plot + Table – one figure per scenario ─────────────────────────────────

SCENARIOS = sorted(agg["scenario"].unique())
MODELS    = ["tabpfn", "distnet"]
MODEL_STYLES = {
    "tabpfn":  {"color": "#1f77b4", "label": "TabPFN"},
    "distnet": {"color": "#ff7f0e", "label": "DistNet"},
}
# Two models → two fixed vertical offsets so labels never share the same slot
MODEL_ANNOT_OFFSET = {
    "tabpfn":  14,   # always above the point
    "distnet": -18,  # always below the point
}
MODEL_ANNOT_VA = {
    "tabpfn":  "bottom",
    "distnet": "top",
}

SAVE_DIR = r"C:\Users\ihagv\Desktop\Study Project UFR\Study_Project_Code\results\plots"
os.makedirs(SAVE_DIR, exist_ok=True)

for scenario in SCENARIOS:
    # ── Plot ──────────────────────────────────────────────────────────────────
    fig, ax = plt.subplots(figsize=(10, 5))

    for model, style in MODEL_STYLES.items():
        sub = agg[(agg["scenario"] == scenario) & (agg["model_name"] == model)].sort_values("context_size")

        if sub.empty:
            print(f"  [WARN] No data for model={model}, scenario={scenario}")
            continue

        x    = sub["context_size"].values
        mean = sub["mean"].values
        std  = sub["std"].values

        ax.plot(x, mean, marker="o", color=style["color"], label=style["label"], linewidth=2, zorder=3)
        ax.fill_between(x, mean - std, mean + std, alpha=0.25, color=style["color"])

        # # One fixed direction per model → no overlap between models
        # y_offset = MODEL_ANNOT_OFFSET[model]
        # va       = MODEL_ANNOT_VA[model]
        # for xi, mi in zip(x, mean):
        #     ax.annotate(
        #         f"{mi:.3f}",
        #         xy=(xi, mi),
        #         xytext=(0, y_offset),
        #         textcoords="offset points",
        #         ha="center",
        #         va=va,
        #         fontsize=7.5,
        #         color=style["color"],
        #         bbox=dict(boxstyle="round,pad=0.15", fc="white", ec="none", alpha=0.6),
        #     )

    ax.set_xscale("log")
    all_sizes = sorted(agg[agg["scenario"] == scenario]["context_size"].unique())
    ax.set_xticks(all_sizes)
    ax.set_xticklabels([str(s) for s in all_sizes], fontsize=9)

    ax.set_title(f"Scenario: {scenario}", fontsize=14, fontweight="bold")
    ax.set_xlabel("Context size (number of samples) — log scale", fontsize=12)
    ax.set_ylabel("NLLH", fontsize=12)
    ax.legend(fontsize=11)
    ax.grid(True, linestyle="--", alpha=0.5, which="both")
    fig.tight_layout()

    save_path = os.path.join(SAVE_DIR, f"scenario_{scenario}.png")
    fig.savefig(save_path, dpi=150)
    plt.show()
    plt.close(fig)
    print(f"Saved plot: {save_path}")

    # ── Table ─────────────────────────────────────────────────────────────────
    table_rows = []
    for model in MODELS:
        sub = agg[(agg["scenario"] == scenario) & (agg["model_name"] == model)].sort_values("context_size")
        if sub.empty:
            continue
        for _, row in sub.iterrows():
            table_rows.append({
                "Model":        MODEL_STYLES[model]["label"],
                "Context Size": int(row["context_size"]),
                "Mean NLLH":    round(row["mean"], 4),
                "Std NLLH":     round(row["std"],  4),
            })

    table_df = pd.DataFrame(table_rows)

    # Pivot for a compact side-by-side view
    pivot = table_df.pivot(index="Context Size", columns="Model", values=["Mean NLLH", "Std NLLH"])
    pivot.columns = [f"{model} – {metric}" for metric, model in pivot.columns]
    pivot = pivot.sort_index().reset_index()

    print(f"\n── Table: Scenario '{scenario}' ──")
    print(pivot.to_string(index=False))
    print()

    # Save table as CSV
    csv_path = os.path.join(SAVE_DIR, f"scenario_{scenario}_table.csv")
    pivot.to_csv(csv_path, index=False)
    print(f"Saved table : {csv_path}\n")

print("Done – all plots and tables saved to:", SAVE_DIR)
```

```python
import os
import pickle
import numpy as np
import pandas as pd
from collections import defaultdict
import plotly.graph_objects as go
from IPython.display import display

# ── 1. Load all .pkl files ────────────────────────────────────────────────────

PKL_DIR = r"C:\Users\ihagv\Desktop\Study Project UFR\Study_Project_Code\results\results_data\tabpfn_context\metadata"

records = []
for fname in os.listdir(PKL_DIR):
    if not fname.endswith(".pkl"):
        continue
    fpath = os.path.join(PKL_DIR, fname)
    with open(fpath, "rb") as f:
        d = pickle.load(f)
    records.append({
        "model_name":   d["model_name"],
        "scenario":     d["scenario"],
        "fold":         d["fold"],
        "context_size": d["context_size"],
        "nllh":         d["result_metrics"]["nllh"],
    })

df = pd.DataFrame(records)
print(f"Loaded {len(df)} records")
print("Scenarios :", sorted(df["scenario"].unique()))
print("Models    :", sorted(df["model_name"].unique()))
print("Context sizes:", sorted(df["context_size"].unique()))

# ── 2. Aggregation ────────────────────────────────────────────────────────────
fold_means = (
    df.groupby(["model_name", "scenario", "fold", "context_size"])["nllh"]
    .mean()
    .reset_index()
    .rename(columns={"nllh": "nllh_fold_mean"})
)

agg = (
    fold_means.groupby(["model_name", "scenario", "context_size"])["nllh_fold_mean"]
    .agg(mean="mean", std="std")
    .reset_index()
)

# ── 3. Config ─────────────────────────────────────────────────────────────────

SCENARIOS = sorted(agg["scenario"].unique())
MODELS    = ["tabpfn", "distnet"]
MODEL_STYLES = {
    "tabpfn":  {"color": "#1f77b4", "label": "TabPFN"},
    "distnet": {"color": "#ff7f0e", "label": "DistNet"},
}

SAVE_DIR = r"C:\Users\ihagv\Desktop\Study Project UFR\Study_Project_Code\results\plots"
os.makedirs(SAVE_DIR, exist_ok=True)

# ── helper: compute per-point text positions ──────────────────────────────────
def get_text_positions(model, x, mean, all_model_data):
    """
    For each point in (x, mean), decide 'top center' or 'bottom center'
    by comparing against the other model's value at the same context_size.
    The higher model at each x goes above, the lower goes below.
    If only one model exists at that x, default to top center.
    """
    other_models = {m: d for m, d in all_model_data.items() if m != model}
    positions = []
    for xi, mi in zip(x, mean):
        other_vals = []
        for other_data in other_models.values():
            mask = other_data["x"] == xi
            if mask.any():
                other_vals.append(other_data["mean"][mask][0])
        if other_vals:
            other_mean = other_vals[0]
            positions.append("top center" if mi >= other_mean else "bottom center")
        else:
            positions.append("top center")
    return positions

# ── 4. Plot + Table per scenario ──────────────────────────────────────────────

for scenario in SCENARIOS:
    fig = go.Figure()

    # collect data for all models first (needed for relative positioning)
    all_model_data = {}
    for model in MODELS:
        sub = agg[(agg["scenario"] == scenario) & (agg["model_name"] == model)].sort_values("context_size")
        if sub.empty:
            continue
        all_model_data[model] = {
            "x":    sub["context_size"].values,
            "mean": sub["mean"].values,
            "std":  sub["std"].values,
        }

    for model in MODELS:
        if model not in all_model_data:
            print(f"  [WARN] No data for model={model}, scenario={scenario}")
            continue

        style = MODEL_STYLES[model]
        x     = all_model_data[model]["x"]
        mean  = all_model_data[model]["mean"]
        std   = all_model_data[model]["std"]

        text_positions = get_text_positions(model, x, mean, all_model_data)

        # ── shaded std band ───────────────────────────────────────────────────
        fig.add_trace(go.Scatter(
            x=np.concatenate([x, x[::-1]]),
            y=np.concatenate([mean + std, (mean - std)[::-1]]),
            fill="toself",
            fillcolor=style["color"],
            opacity=0.15,
            line=dict(width=0),
            hoverinfo="skip",
            showlegend=False,
        ))

        # ── main line + markers + labels ─────────────────────────────────────
        # Plotly requires a single textposition per trace, so split into
        # individual annotation shapes to allow per-point positioning
        fig.add_trace(go.Scatter(
            x=x,
            y=mean,
            mode="lines+markers",
            name=style["label"],
            line=dict(color=style["color"], width=2.5),
            marker=dict(size=8, color=style["color"],
                        line=dict(width=1.5, color="white")),
            hovertemplate=(
                f"<b>{style['label']}</b><br>"
                "Context size: %{x}<br>"
                "Mean NLLH: %{y:.4f}<br>"
                f"Std: %{{customdata:.4f}}<extra></extra>"
            ),
            customdata=std,
        ))

        # ── per-point annotations via layout.annotations ──────────────────────
        for xi, mi, pos in zip(x, mean, text_positions):
            ay = -28 if pos == "top center" else 28
            fig.add_annotation(
                x=np.log10(xi),   # log-axis requires log10 value for x
                y=mi,
                xref="x",
                yref="y",
                text=f"<b>{mi:.3f}</b>",
                showarrow=True,
                arrowhead=0,
                arrowwidth=1,
                arrowcolor=style["color"],
                ax=0,
                ay=ay,
                font=dict(size=10, color=style["color"]),
                bgcolor="white",
                bordercolor=style["color"],
                borderwidth=1,
                borderpad=3,
                opacity=0.9,
            )

    all_sizes = sorted(agg[agg["scenario"] == scenario]["context_size"].unique())

    fig.update_layout(
        title=dict(
            text=f"Scenario: <b>{scenario}</b>",
            font=dict(size=18),
            x=0.5,
            xanchor="center",
        ),
        xaxis=dict(
            title="Context size (number of samples)",
            type="log",
            tickvals=all_sizes,
            ticktext=[str(s) for s in all_sizes],
            tickfont=dict(size=11),
            gridcolor="#e0e0e0",
            showline=True,
            linecolor="#aaaaaa",
            mirror=True,
        ),
        yaxis=dict(
            title="NLLH",
            gridcolor="#e0e0e0",
            showline=True,
            linecolor="#aaaaaa",
            mirror=True,
        ),
        legend=dict(
            bgcolor="rgba(255,255,255,0.85)",
            bordercolor="#cccccc",
            borderwidth=1,
            font=dict(size=12),
        ),
        plot_bgcolor="white",
        paper_bgcolor="white",
        hovermode="x unified",
        width=950,
        height=560,
        margin=dict(l=60, r=40, t=70, b=60),
    )

    fig.show()

    html_path = os.path.join(SAVE_DIR, f"scenario_{scenario}.html")
    png_path  = os.path.join(SAVE_DIR, f"scenario_{scenario}.png")
    fig.write_html(html_path)
    fig.write_image(png_path, scale=2)
    print(f"Saved: {html_path}")
    print(f"Saved: {png_path}")

    # ── Table ─────────────────────────────────────────────────────────────────
    table_rows = []
    for model in MODELS:
        sub = agg[(agg["scenario"] == scenario) & (agg["model_name"] == model)].sort_values("context_size")
        if sub.empty:
            continue
        for _, row in sub.iterrows():
            table_rows.append({
                "Model":        MODEL_STYLES[model]["label"],
                "Context Size": int(row["context_size"]),
                "Mean NLLH":    round(row["mean"], 4),
                "Std NLLH":     round(row["std"],  4),
            })

    table_df = pd.DataFrame(table_rows)
    pivot = table_df.pivot(index="Context Size", columns="Model", values=["Mean NLLH", "Std NLLH"])
    pivot.columns = [f"{model} – {metric}" for metric, model in pivot.columns]
    pivot = pivot.sort_index().reset_index()

    print(f"\n── Table: Scenario '{scenario}' ──")
    display(pivot.style
            .format(precision=4)
            .set_caption(f"Scenario: {scenario}")
            .set_table_styles([
                {"selector": "th", "props": [("background-color", "#f0f4ff"),
                                             ("font-weight", "bold"),
                                             ("text-align", "center")]},
                {"selector": "td", "props": [("text-align", "center")]},
            ])
    )

    csv_path = os.path.join(SAVE_DIR, f"scenario_{scenario}_table.csv")
    pivot.to_csv(csv_path, index=False)
    print(f"Saved table: {csv_path}\n")

print("Done – all plots and tables saved to:", SAVE_DIR)
```
